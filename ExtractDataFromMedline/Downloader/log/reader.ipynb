```python
 llist = []
with open('2017.log','r') as fin:
    for line in fin:
        if len(line) > 100:
            #print(line)

            line = line.split(',')[1].split('..')[1]
            fileName,pmid = line.split('.xml')[0].split('/')[3],line.split('.xml')[1]
            llist.append((fileName,pmid))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data=llist, columns=['fileName','pmID'])
```
