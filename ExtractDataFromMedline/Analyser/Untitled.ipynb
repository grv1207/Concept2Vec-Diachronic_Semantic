```python
llist = []
with open('log1','r') as fout:
    for line in fout:
        if line.__contains__(': 30000'):
            furl,status =line.split(':')[0],line.split(':')[1].strip('\n')
            llist.append((furl,status))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data=llist)
```
